```python
import csv

def import_data_from_csv(file_path):
    data = []
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        for row in csv_reader:
            data.append(row)
    return data

```

```python
def calculate_average(lst):
    if len(lst) == 0:
        return 0  # Handle empty list case to avoid division by zero
    else:
        return sum(lst) / len(lst)
    
    
```

```python
import matplotlib.pyplot as plt
# Example usage
csv_file_path = 'iou_log_refcoco.csv'
imported_data = import_data_from_csv(csv_file_path)
scores = [float(sublist[3]) for sublist in imported_data]
enum = [i for i in range(len(scores))]

avg_list = []
for i in range(len(scores)):
    sublist = scores[0:i]
    avg_list.append(calculate_average(sublist)*100)

plt.plot(enum, avg_list)

# Adding labels and title
plt.xlabel('num_images')
plt.ylabel('average score')
plt.title('Evaluation graph for RefCOCO')

# Displaying the plot
plt.show()
print("Min score: ",min(scores))
print("Max score: ",max(scores))
print("Final avg: ", avg_list[-1])
```
